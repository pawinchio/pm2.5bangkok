```python
import pandas as pd
```

```python
pm = pd.read_csv('bangkok_pm25.csv')
pm
```

```python
pm2=pm.groupby('Year').mean()
pm2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(pm2['Year'],pm2['PM2.5'], 'r') # 'r' is the color red
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.title('pm per Year')
plt.show()
```
